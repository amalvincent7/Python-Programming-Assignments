### Q1. Write a Python Program to Find the Factorial of a Number?

```python
number = int(input("Enter a number for factorial value : "))

a=1
for i in range(1,number+1):
    a=a*i
print("Factorial of {num} is {a}".format(num=number,a=a))
```

### Q2. Write a Python Program to Display the multiplication Table?

```python
number = int(input("Enter the number for multiplication table : "))

i=1
print("\n")
while i<11:
    mult=i*number
    print("{a} x {b} = {c}".format(a=i,b=number,c=mult),"\n")
    i+=1
    
```

### Q3. Write a Python Program to Print the Fibonacci sequence?

```python
upper_limit=100

seq=[]
for i in range(0,upper_limit+1):
    if i==0 or i==1:
        seq.append(i)
    else:
        next_num=seq[(i-2)]+seq[(i-1)]
        seq.append(next_num)
        
for i in seq:
    print(i)
```

### Q4. Write a Python Program to Check Armstrong Number?

```python
number=input("Enter a number to check armstrong condition : ")

sum=0
for i in number:
    a=int(i)**len(number)
    sum+=a
    
if sum==int(number):
    print("Number is a Armstrong number")
else:
    print("Number is not a Armstrong number")
```

### Q5. Write a Python Program to Find Armstrong Number in an Interval?

```python
lower_limit=int(input("Enter a lower limit number to check armstrong condition : "))
upper_limit=int(input("Enter a upper limit number to check armstrong condition : "))

for i in range(lower_limit,upper_limit+1):
    sum=0
    for j in str(i):
        a=int(j)**len(str(i))
        sum+=a
        
    if sum==i:
        print(i)
```

### Q6. Write a Python Program to Find the Sum of Natural Numbers?

```python
lower_limit=int(input("Enter a lower limit : "))
upper_limit=int(input("Enter a upper limit : "))

sum=0
for i in range(lower_limit,upper_limit+1):
    if i>0:
        sum+=i
    else:
        continue
print("Sum of natural numbers in given limit is",sum)
    
```
